```python
from sklearn.cluster import KMeans
```

```python
kmeans1=KMeans(n_clusters=4,random_state=2)
y_pred=kmeans1.fit_predict(data[['petal_length','petal_width']])
y_pred
```

```python
data_with_cluster1=data.copy()
data_with_cluster1['cluster1']=y_pred
data_with_cluster1
```

```python
centroids=kmeans1.cluster_centers_
centroids
```

```python
data_with_cluster1['species']=data_with_cluster1['species'].map({'Iris_setosa':0,'Iris-versicolor':1,'Iris-verginica':2})
```

```python
kmeans2=KMeans(n_clusters=5,random_state=2)
y_pred1=kmeans2.fit_predict(data[['sepal_length','sepal_width']])
y_pred1
```

```python
data_with_cluster2=data.copy()
data_with_cluster2['cluster2']=y_pred1
data_with_cluster2
```

```python
plt.scatter(data_with_cluster2['sepal_length'],data_with_cluster2['sepal_width'],c=data_with_cluster2["cluster2"],cmap='rainbow')
```
